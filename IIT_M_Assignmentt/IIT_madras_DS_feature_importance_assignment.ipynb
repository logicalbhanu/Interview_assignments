#### Preprocessing the dataframe

```python
from sklearn.preprocessing import OneHotEncoder

def one_hot_creator(df, attribute, list_of_new_attributes):
    #creating instance of one-hot-encoder
    encoder = OneHotEncoder(handle_unknown='ignore')
    
    #perform one-hot encoding on 'team' column 
    encoder_df = pd.DataFrame(encoder.fit_transform(df[[attribute]]).toarray())
    
    #merge one-hot encoded columns back with original DataFrame
    final_df = df.join(encoder_df)
    
    new_columns = {i:list_of_new_attributes[i] for i in range(len(list_of_new_attributes))} 
    final_df.rename(columns = new_columns, inplace=True) 
    final_df.drop(attribute,axis=1, inplace=True)
    
    #view final df
    return final_df
```

```python
new_df = X
```

```python
new_df = one_hot_creator(new_df,'weekday', ['sun','mon','tue','wed','thu','fri','sat'])
new_df = one_hot_creator(new_df,'season', ['season_1','season_2','season_3','season_4'])
new_df = one_hot_creator(new_df, 'mnth', ['jan','feb','mar','apr','may','jun','jul','aug',
                                          'sep','oct','nov','dec']) 
new_df = one_hot_creator(new_df, 'weathersit',['weathersit_1','weathersit_2','weathersit_3'] )
```

```python
new_df.drop(['instant','dteday'],axis=1, inplace=True)
```

#### standardising the data

```python
for i in ['temp','atemp','hum','windspeed','casual','registered']:
    new_df[i] = (new_df[i]-new_df[i].mean())/new_df[i].std() # normalized_df=(df-df.mean())/df.std()
```

```python
y_new = (y-y.mean()/y.std())
```

#### Creating Linear Regression model

```python
import numpy as np
from sklearn.linear_model import LinearRegression
reg = LinearRegression().fit(new_df, y_new)

print('coef : ',np.abs(reg.coef_))

print('intercept : ',reg.intercept_)
```

#### Extracting important features

```python
feat_imp = np.argsort(np.abs(reg.coef_))[-1::-1]
```

```python
new_df.columns[feat_imp]
```

#### Top 3 important features 

```python
top_3_imp_features = new_df.columns[feat_imp][:3]
print('Top three features : ', top_3_imp_features.values)
```
